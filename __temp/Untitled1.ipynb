```python
'''
=================
3D wireframe plot
=================

A very basic demonstration of a wireframe plot.
'''

from mpl_toolkits.mplot3d import axes3d
import matplotlib.pyplot as plt
import math
import numpy as np
import csv
from matplotlib import cm
import matplotlib as mpl


fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
X = []
Y = []
Z = []
A = []
B = []
C = []
#with Prop delay
for x in range(4,21):
    csvfile = open("2node/data21_t"+str(x),"r")
    readernya = csv.reader(csvfile,delimiter=",")
    dataset = []
    for r in readernya:
        data = []
        data.append(int(r[0]))
        data.append(float(r[2]))
        data.append(round(float(r[4])))
        dataset.append(data)

    dataset = np.array(dataset)
    X.append([i[0] for i in dataset[:,0:1].tolist()])
    Y.append([i[0] for i in dataset[:,1:2].tolist()])
    Z.append([i[0] for i in dataset[:,2:3].tolist()])

X = np.array(X)
Y = np.array(Y)
Z = np.array(Z)

# Plot a basic wireframe.
#ax.plot_wireframe(X, Y, Z)
ax.set_xlabel('State')
ax.set_ylabel('Lambda')
ax.set_zlabel('Waktu Komputasi (ms)')
ax.plot_surface(X, Y, Z, color='none', edgecolor='crimson', label='with prop delay')

#with no prop delay
for x in range(4,21):
    csvfile = open("2node/data21_t"+str(x),"r") #data21_t4, data21_t4, dst..
    readernya = csv.reader(csvfile,delimiter=",")
    dataset = []
    for r in readernya:
        data = []
        data.append(int(r[0]))
        data.append(float(r[2]))
        data.append(round(float(r[3])))
        dataset.append(data)

    dataset = np.array(dataset)
    A.append([i[0] for i in dataset[:,0:1].tolist()])
    B.append([i[0] for i in dataset[:,1:2].tolist()])
    C.append([i[0] for i in dataset[:,2:3].tolist()])

A = np.array(A)
B = np.array(B)
C = np.array(C)

# Plot a basic wireframe.
#ax.plot_wireframe(X, Y, Z)
ax.set_xlabel('State')
ax.set_ylabel('Lambda*t')
ax.set_zlabel('Waktu Komputasi (ms)')
ax.plot_surface(A, B, C, color='none', edgecolor='steelblue', label='with non prop delay')
fake2Dline1 = mpl.lines.Line2D([0],[0], linestyle="none", c='crimson', marker = 'o')
fake2Dline2 = mpl.lines.Line2D([0],[0], linestyle="none", c='steelblue', marker = 'o')
ax.legend([fake2Dline1,fake2Dline2], ['with prop delay','with non prop delay'], numpoints = 1)
plt.show()

''' cyclic'''
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
X = []
Y = []
Z = []
A = []
B = []
C = []
#with Prop delay
for x in range(4,8):
    csvfile = open("2node/data21_c"+str(x),"r")
    readernya = csv.reader(csvfile,delimiter=",")
    dataset = []
    for r in readernya:
        data = []
        data.append(int(r[0]))
        data.append(float(r[2]))
        data.append(round(float(r[4])))
        dataset.append(data)

    dataset = np.array(dataset)
    X.append([i[0] for i in dataset[:,0:1].tolist()])
    Y.append([i[0] for i in dataset[:,1:2].tolist()])
    Z.append([i[0] for i in dataset[:,2:3].tolist()])

X = np.array(X)
Y = np.array(Y)
Z = np.array(Z)

# Plot a basic wireframe.
#ax.plot_wireframe(X, Y, Z)
ax.set_xlabel('State')
ax.set_ylabel('Lambda')
ax.set_zlabel('Waktu Komputasi (ms)')
ax.plot_surface(X, Y, Z, color='none', edgecolor='crimson', label='with prop delay')

#with no prop delay
for x in range(4,8):
    csvfile = open("2node/data21_c"+str(x),"r")
    readernya = csv.reader(csvfile,delimiter=",")
    dataset = []
    for r in readernya:
        data = []
        data.append(int(r[0]))
        data.append(float(r[2]))
        data.append(round(float(r[3])))
        dataset.append(data)

    dataset = np.array(dataset)
    A.append([i[0] for i in dataset[:,0:1].tolist()])
    B.append([i[0] for i in dataset[:,1:2].tolist()])
    C.append([i[0] for i in dataset[:,2:3].tolist()])

A = np.array(A)
B = np.array(B)
C = np.array(C)

# Plot a basic wireframe.
#ax.plot_wireframe(X, Y, Z)
ax.set_xlabel('State')
ax.set_ylabel('Lambda*t')
ax.set_zlabel('Waktu Komputasi (ms)')
ax.plot_surface(A, B, C, color='none', edgecolor='steelblue', label='with non prop delay')
fake2Dline1 = mpl.lines.Line2D([0],[0], linestyle="none", c='crimson', marker = 'o')
fake2Dline2 = mpl.lines.Line2D([0],[0], linestyle="none", c='steelblue', marker = 'o')
ax.legend([fake2Dline1,fake2Dline2], ['with prop delay','with non prop delay'], numpoints = 1)
plt.show()
```
